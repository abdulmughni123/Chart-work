```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
    weather=pd.read_csv("https://raw.githubusercontent.com/alanjones2/dataviz/master/london2018.csv")
    weather

    weather.plot(x="Month",y="Tmin",color="black",marker="o",markerfacecolor="red")
    weather.plot.bar(x="Month",y="Tmin",color="black")

    weather.plot.bar(x="Month",y=["Tmin","Sun"],color=["Red","Green"])
    weather.plot.scatter(x="Month",y="Tmin",color="purple")

    weather.plot.bar(x="Year",y=["Month","Sun","Rain","Tmin"],color=["pink","orange","purple","red","yellow","blue"])
```
